```python
import os
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import cv2
```

```python
categories=['with_mask','without_mask']
```

```python
data=[]
for category in categories:
    path=os.path.join('train',category)
    label=categories.index(category)
    
    for file in os.listdir(path):
        
        img_path=os.path.join(path,file)
        img=cv2.imread(img_path)
        img=cv2.resize(img,(224,224))
        
        data.append([img,label])
```

```python
len(data)
```

```python
import random
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]

for features,label in data:
    x.append(features)
    y.append(label)
```

```python
len(x)
```

```python
len(y)
```

```python
import numpy as np
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
x=x/255
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16()
```

```python
vgg.summary()
```

```python
from tensorflow.keras import Sequential
```

```python
model=Sequential()
```

```python
models=vgg.layers[:-1]
for layer in models:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from tensorflow.keras.layers import Dense,Conv2D,BatchNormalization,MaxPooling2D,Flatten,Dropout
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=5,validation_data=(X_test,Y_test),batch_size=16)
```

```python
def detect_face_mask(img):
    y_pred=model.predict(img.reshape(1,224,224,3))
    return y_pred[0][0]
```

```python
def draw_label(img,text,pos,bg_color):
    text_size=cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    end_x=pos[0]+text_size[0][0]+2
    end_y=pos[0]+text_size[0][1]-2

    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
```

```python
haar=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    coords=haar.detectMultiScale(img)
    return coords
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,frame=cap.read()
    #Detection function
    frame=cv2.flip(frame,1)
    img=cv2.resize(frame,(224,224))
    y_pred=detect_face_mask(img)
    coords=detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    for x,y,w,h in coords:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    if(round(y_pred)>0.5):
        draw_label(frame,"No Mask",(25,25),(0,0,255))
    else:
        draw_label(frame,"Mask",(25,25),(0,255,0))
    
    cv2.imshow("window",frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
#For Face Detection
while True:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    coords=detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    for x,y,w,h in coords:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    cv2.imshow("window",frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
